# Streaming Platforms Datasets Preprocessing

## Titles preprocessing

```python
import pandas as pd
```

```python
titles_amazon = pd.read_csv('/content/titles_amazon.csv')
titles_amazon.head()
```

```python
titles_disney = pd.read_csv('/content/titles_disney.csv')
titles_disney.head()
```

```python
titles_hulu = pd.read_csv('/content/titles_hulu.csv')
titles_hulu.head()
```

```python
titles_netflix = pd.read_csv('/content/titles_netflix.csv')
titles_netflix.head()
```

```python
titles = pd.concat([titles_netflix, titles_amazon, titles_hulu, titles_disney], ignore_index=True)
titles
```

```python
titles.info()
```

```python
titles[titles['age_certification'].isna()].head()
```

```python
print(f"There are {len(titles[titles['age_certification'].isna()])} movies and tv shows which are not rated for age restrictions.")
```

Let's replace these `null` values with `Not rated`:

```python
titles['age_certification'].fillna('Not rated', inplace=True)
```

Even though columns `genres` and `production_countries` don't have `null` values, there are some rows with unspecified genres and production countries accordingly, which looks like `[]`. Let's replace these values to `unspecified`:

```python
titles['genres'].replace('[]', 'unspecified', inplace=True)
```

```python
titles['production_countries'].replace('[]', 'unspecified', inplace=True)
```

```python
print(f"In the column 'seasons' there are {len(titles[titles['seasons'].isna()])} null values.")
```

Let's see how many of these `null` values are related to tv shows:

```python
titles[titles['seasons'].isna()]['type'].value_counts()
```

We can see that all of these `null` values are only related to movies. In this case we can replace them with `0`, since movies don't have seasons:

```python
titles['seasons'].fillna(0, inplace=True)
```

```python
print(f"There are also {len(titles[titles['imdb_score'].isna()])} null values in the 'imdb_score' column and {len(titles[titles['imdb_votes'].isna()])} null values in the 'imdb_votes' column.")
```

Let's replace them with `0`:

```python
titles['imdb_score'].fillna(0, inplace=True)
titles['imdb_votes'].fillna(0, inplace=True)
```

We are not planning to use `imdb_id` column, but anyway let's replace `null` values in this column with `Not rated`:

```python
titles['imdb_id'].fillna('Not rated', inplace=True)
```

```python
print(f"There are {titles.duplicated().sum()} duplicated records in the dataset.")
```

```python
titles.info()
```

```python
print(f"In the column 'description' there are {len(titles[titles['description'].isna()])} null values.")
```

But we will leave these `null` values in the `description` column as it is, it is not critical in this case.

Let's remove columns we are not planning to use: `tmdb_popularity` and `tmdb_score`:

```python
titles = titles.drop(['tmdb_popularity','tmdb_score'], axis=1)
titles.head()
```

```python
titles.to_csv('titles_all.csv', index=False)
```

```python
print(f"In the column 'character' there are {len(credits[credits['character'].isna()])} null values.")
```
